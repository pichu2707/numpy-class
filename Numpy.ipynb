## DotProducts

```python
a = np.array([1,2])
b = np.array([2,3])

dot2 = 0
for e,f in zip(a,b):
  dot2 += e*f

print('a: ',a)
print('b: ',b)
print('a*b: ',a*b)
print('dot1: ',a.dot(b))
print('dot2: ',dot2)
print('np.dot(a,b):',np.dot(a,b))
print('np.dot(b,a):',np.dot(b,a))
```

## Código de ejemplo para usar estas medias y desviación

```python
from numpy import percentile
from numpy.random import rand

#Generar un dato random
data = np.random.rand(1000)
print(f'datos de data: {data}')
#calculando los cuartile mínmos y mámiximos

cuartiles = percentile(data, [25,50,75])
data_min, data_max = data.min(), data.max()

#Impriminedo datos
print('Mínimo: ',  data_min)
print('Q1: %.3f'% cuartiles[0])
print('Medio: %.3f' % cuartiles[1])
print('Mean Val: %.3f'% data.mean())
print('std: %.3f'% data.std())
print('Q3: %.3f'% cuartiles[2])
print('Máximo: %.3f'% data_max)
```

###¿Qué es la regresión lineal?:
La ecuación en un hipeplano con un grado menor de la dimensionalidad que tendía el conjunto de los datos.
Tenemos datos que estén en 3D ya no es un hiperplano si no que pasa a ser un plano.

La regresión lineal es adecuada cunado los puntos del conjunto están distribuidas de tal forma que pueden aproximarse de manerta razonable mediante hiperplano. Si no es aśi, intenamos ajustar otro tipos de superficies a los puntos del conjunto de datos.
